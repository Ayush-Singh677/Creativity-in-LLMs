```python
from transformers import GPT2LMHeadModel, GPT2TokenizerFast
from datasets import load_dataset
import torch
from tqdm import tqdm

# Function to calculate perplexity for a given model and tokenizer
def calculate_perplexity(model, tokenizer, dataset_split, device, stride=50):
    max_length = model.config.n_positions
    data = dataset_split["text"]
    encodings = tokenizer("\n\n".join(data), return_tensors="pt")
    seq_len = encodings.input_ids.size(1)

    nlls = []
    prev_end_loc = 0
    for begin_loc in tqdm(range(0, seq_len, stride), desc="Calculating Perplexity"):
        end_loc = min(begin_loc + max_length, seq_len)
        trg_len = end_loc - prev_end_loc
        input_ids = encodings.input_ids[:, begin_loc:end_loc].to(device)
        target_ids = input_ids.clone()
        target_ids[:, :-trg_len] = -100

        with torch.no_grad():
            outputs = model(input_ids, labels=target_ids)
            neg_log_likelihood = outputs.loss

        nlls.append(neg_log_likelihood)
        prev_end_loc = end_loc
        if end_loc == seq_len:
            break

    return torch.exp(torch.stack(nlls).mean())

# Initialize device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load pretrained GPT-2 model and tokenizer
pretrained_model_id = "gpt2-large"
pretrained_model = GPT2LMHeadModel.from_pretrained(pretrained_model_id).to(device)
pretrained_tokenizer = GPT2TokenizerFast.from_pretrained(pretrained_model_id)

# Load the dataset
dataset = load_dataset("Ayush-Singh/jokes-sample-3")

# Splits to evaluate
splits = ["train_sample","test", "random"]

# Calculate perplexity for each split
for split in splits:
    print(f"\nCalculating perplexity for pretrained model on the '{split}' split...")
    dataset_split = dataset[split]
    pretrained_ppl = calculate_perplexity(pretrained_model, pretrained_tokenizer, dataset_split, device)
    print(f"Pretrained model perplexity on {split}: {pretrained_ppl:.2f}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from datasets import load_dataset

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load dataset
dataset = load_dataset("Ayush-Singh/jokes-sample-3")

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained('gpt2')
model = AutoModelForCausalLM.from_pretrained('gpt2').to(device)

# Set the EOS token as the padding token
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Tokenize the dataset
def tokenize_function(examples):
    inputs =  tokenizer(examples['text'], truncation=True, padding='max_length', max_length=128)
    inputs['labels'] = inputs['input_ids'].copy()
    return inputs


tokenized_datasets = dataset.map(tokenize_function, batched=True)
print(tokenized_datasets)
from datasets import load_dataset

# Load your dataset (or use the one you already have)
# Assuming `dataset` is your DatasetDict
# dataset = load_dataset(...)

# View one row from the 'test' dataset
test_row = tokenized_datasets['test'][0]  # Index 0 for the first row
print(test_row)
```

```python
# Define training arguments
training_args = TrainingArguments(
    output_dir='/content/results',
    evaluation_strategy='epoch',
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='/logs'
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test'],
)

# Train the model
trainer.train()

# save the model and tokenizer explicitly
model_output_dir = '/content/model/gpt2_jokes'

model.save_pretrained(model_output_dir)
tokenizer.save_pretrained(model_output_dir)
```

```python
from transformers import GPT2LMHeadModel, GPT2TokenizerFast
from datasets import load_dataset
import torch
from tqdm import tqdm

# Function to calculate perplexity for a given model and tokenizer
def calculate_perplexity(model, tokenizer, dataset_split, device, stride=50):
    max_length = model.config.n_positions
    data = dataset_split["text"]
    encodings = tokenizer("\n\n".join(data), return_tensors="pt")
    seq_len = encodings.input_ids.size(1)

    nlls = []
    prev_end_loc = 0
    for begin_loc in tqdm(range(0, seq_len, stride), desc="Calculating Perplexity"):
        end_loc = min(begin_loc + max_length, seq_len)
        trg_len = end_loc - prev_end_loc
        input_ids = encodings.input_ids[:, begin_loc:end_loc].to(device)
        target_ids = input_ids.clone()
        target_ids[:, :-trg_len] = -100

        with torch.no_grad():
            outputs = model(input_ids, labels=target_ids)
            neg_log_likelihood = outputs.loss

        nlls.append(neg_log_likelihood)
        prev_end_loc = end_loc
        if end_loc == seq_len:
            break

    return torch.exp(torch.stack(nlls).mean())

# Initialize device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load fine-tuned model and tokenizer
fine_tuned_model_path = "/content/model/gpt2_jokes"
fine_tuned_model = GPT2LMHeadModel.from_pretrained(fine_tuned_model_path).to(device)
fine_tuned_tokenizer = GPT2TokenizerFast.from_pretrained(fine_tuned_model_path)

# Load pretrained model and tokenizer
pretrained_model_id = "gpt2"
pretrained_model = GPT2LMHeadModel.from_pretrained(pretrained_model_id).to(device)
pretrained_tokenizer = GPT2TokenizerFast.from_pretrained(pretrained_model_id)

# Load the dataset
dataset = load_dataset("Ayush-Singh/jokes-sample-3")
splits = [ "train_sample","test", "random"]  # Dataset splits to evaluate

# Evaluate perplexity for both models on all splits
results = {}

for split in splits:
    print(f"\nEvaluating split: {split}")

    # Evaluate fine-tuned model
    print("Calculating perplexity for fine-tuned model...")
    fine_tuned_ppl = calculate_perplexity(fine_tuned_model, fine_tuned_tokenizer, dataset[split], device)
    print(f"Fine-tuned model perplexity on {split}: {fine_tuned_ppl:.2f}")

    # Evaluate pretrained model
    print("Calculating perplexity for pretrained model...")
    pretrained_ppl = calculate_perplexity(pretrained_model, pretrained_tokenizer, dataset[split], device)
    print(f"Pretrained model perplexity on {split}: {pretrained_ppl:.2f}")

    # Store results
    results[split] = {
        "fine_tuned": fine_tuned_ppl.item(),
        "pretrained": pretrained_ppl.item()
    }

# Print summary of results
print("\nSummary of Perplexity Results:")
for split, result in results.items():
    print(f"- {split.capitalize()} Split:")
    print(f"  Fine-tuned Model Perplexity: {result['fine_tuned']:.2f}")
    print(f"  Pretrained Model Perplexity: {result['pretrained']:.2f}")
```
